7. Write a Pandas program to find out the alcohol consumption details in the year '1987' or '1989' from the world alcohol consumption dataset.

8. Write a Pandas program to find out the alcohol consumption details by the 'Americas' in the year '1985' from the world alcohol consumption dataset.

```python

filtered_data = df[(df['Year'] == 1985) | (df['WHO region'] == 'Americas')]


print("Alcohol Consumption Details in the Americas in the Year 1985:")
filtered_data
```

Write a Pandas program to find out the records where consumption of beverages per person average >=5 and Beverage Types is Beer from world alcohol consumption dataset.


```python
filtered_data = df[(df['Display Value'] >= 5) & (df['Beverage Types'] == 'Beer')]

print("Records where consumption of beverages per person average >= 5 and Beverage Types is Beer:")
filtered_data
```

10. Write a Pandas program to find out the records where consumption of beverages per person average >=4 and Beverage Types is Beer, Wine, Spirits from world alcohol consumption dataset.

```python
filtered_data = df[(df['Display Value'] >= 4) & (df['Beverage Types'].isin(['Beer', 'Wine', 'Spirits']))]


print("Records where consumption of beverages per person average >= 4 and Beverage Types are Beer, Wine, or Spirits:")
filtered_data
```

11. Write a Pandas program to filter those records where WHO region contains "Ea" substring from world alcohol consumption dataset.
